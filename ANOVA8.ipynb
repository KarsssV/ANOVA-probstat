# **ANOVA**

## Pengantar ANOVA (Analysis of Variance)

**ANOVA (Analysis of Variance)** adalah metode statistik yang digunakan untuk menguji apakah terdapat perbedaan **rata-rata** dari dua kelompok atau lebih. ANOVA sangat berguna dalam **eksperimen atau penelitian komparatif** di mana kita ingin mengetahui pengaruh suatu variabel terhadap hasil atau outcome.

### Intuisi Dasar

Alih-alih membandingkan kelompok satu per satu (dengan uji-t yang berisiko meningkatkan Type I Error), ANOVA memungkinkan kita menguji **semua kelompok secara bersamaan**, kemudian dilanjutkan (jika perlu) dengan **post-hoc test** seperti Tukey HSD untuk mengetahui kelompok mana yang berbeda.

### Asumsi ANOVA

1. **Independensi**: Data antar kelompok bersifat independen.
2. **Normalitas**: Data dalam masing-masing kelompok terdistribusi normal.
3. **Homoskedastisitas**: Varians antar kelompok relatif sama.

---

## Jenis-Jenis ANOVA

### 1. One-Way ANOVA (ANOVA Satu Arah)

#### Deskripsi:
Digunakan ketika kita ingin menguji pengaruh **satu variabel kategorikal (faktor)** terhadap satu variabel numerik (respon).

#### Contoh:
Menguji apakah **jenis pupuk** (Organik, Kimia, Campuran) memengaruhi **tinggi tanaman**.

---

### 2. Two-Way ANOVA (ANOVA Dua Arah)

#### Deskripsi:
Digunakan ketika kita ingin menguji pengaruh **dua variabel kategorikal (faktor A dan B)** terhadap satu variabel numerik.

#### Contoh:
Mengukur pengaruh **jenis pupuk** dan **jenis tanaman** terhadap **tinggi tanaman**.

#### Interaksi:
Two-way ANOVA juga bisa mengevaluasi **interaksi antara dua faktor** — apakah efek salah satu faktor tergantung pada level faktor lainnya.

#### Model:

---

### 3. Repeated Measures ANOVA

#### Deskripsi:
Digunakan ketika **pengukuran dilakukan berulang pada subjek yang sama**, misalnya pengukuran mingguan terhadap pasien.

#### Contoh:
Mengukur tingkat stres seseorang setelah terapi selama 3 minggu berturut-turut.

#### Perbedaan dengan One-Way ANOVA:
- Subjek **tidak independen** antar waktu → korelasi perlu diperhitungkan.
- Biasanya memerlukan struktur data **long format**.

---

### 4. MANOVA (Multivariate ANOVA)

#### Deskripsi:
Perluasan dari ANOVA ketika terdapat **lebih dari satu variabel dependen**.

#### Contoh:
Mengukur pengaruh jenis diet terhadap **berat badan** dan **kadar kolesterol** secara bersamaan.

---

## Catatan Tambahan

| Tipe ANOVA         | Faktor Independen | Faktor Interaksi | Variabel Dependen | Repeated Measure |
|--------------------|-------------------|------------------|-------------------|------------------|
| One-Way ANOVA      | 1                 | ❌               | 1                 | ❌               |
| Two-Way ANOVA      | 2                 | Opsional         | 1                 | ❌               |
| Repeated ANOVA     | 1 (atau lebih)    | ❌               | 1                 | ✅               |
| MANOVA             | ≥1                | Opsional         | ≥2                | ❌               |

Soal ini meminta kita untuk menguji hipotesis mengenai perbedaan rata-rata waktu penyembuhan akibat tiga merek obat sakit kepala yang berbeda. Kita akan menggunakan analisis varians (ANOVA) satu arah karena kita membandingkan rata-rata dari tiga kelompok independen dan diasumsikan bahwa populasi berdistribusi normal dan homogen.

Data dari Gambar:

Obat A: 5, 4, 8, 6, 3, 3, 5, 2

Obat B: 9, 7, 8, 6, 9, 3, 7, 4, 1

Obat C: 7, 6, 9, 4, 7, 2, 3, 4

Pernyataan Masalah:

Kita perlu menguji hipotesis bahwa rata-rata waktu penyembuhan yang diakibatkan oleh ketiga merek obat tersebut tidak berbeda. Kita diberikan asumsi bahwa ketiga populasi berdistribusi normal dan homogen, serta tingkat signifikansi 5%.

Hipotesis:

Hipotesis Nol ( H0 ): Rata-rata waktu penyembuhan dari ketiga kelompok obat adalah sama. ( μA=μB=μC )

Hipotesis Alternatif ( H1 ): Setidaknya ada satu rata-rata waktu penyembuhan yang berbeda dari yang lain.
Langkah-langkah penyelesaian dengan Python:

1. Impor pustaka yang diperlukan: scipy.stats untuk melakukan ANOVA.
2. Masukkan data.
3. Lakukan uji ANOVA satu arah.
4. Interpretasikan hasilnya berdasarkan nilai p (p-value) dan tingkat signifikansi.

```python
import scipy.stats as stats
import pandas as pd # Pustaka ini opsional, tidak digunakan untuk perhitungan utama tapi berguna jika Anda ingin mengelola data lebih lanjut

# Data dari gambar
obat_A = [5, 4, 8, 6, 3, 3, 5, 2]
obat_B = [9, 7, 8, 6, 9, 3, 7, 4, 1]
obat_C = [7, 6, 9, 4, 7, 2, 3, 4]

# Lakukan uji ANOVA satu arah
# Fungsi f_oneway akan mengembalikan statistik F dan nilai p (p-value)
f_statistic, p_value = stats.f_oneway(obat_A, obat_B, obat_C)

# Tingkat signifikansi
alpha = 0.05

print(f"Statistik F: {f_statistic:.4f}")
print(f"Nilai P (p-value): {p_value:.4f}")
print(f"Tingkat Signifikansi (alpha): {alpha}")

# Bandingkan nilai p dengan alpha untuk membuat keputusan
if p_value < alpha:
    print("\nKeputusan: Tolak Hipotesis Nol.")
    print("Kesimpulan: Terdapat perbedaan yang signifikan secara statistik dalam rata-rata waktu penyembuhan antara ketiga merek obat sakit kepala.")
else:
    print("\nKeputusan: Gagal menolak Hipotesis Nol.")
    print("Kesimpulan: Tidak ada perbedaan yang signifikan secara statistik dalam rata-rata waktu penyembuhan antara ketiga merek obat sakit kepala.")

# Untuk melihat rata-rata setiap kelompok (opsional)
mean_A = sum(obat_A) / len(obat_A)
mean_B = sum(obat_B) / len(obat_B)
mean_C = sum(obat_C) / len(obat_C)

print(f"\nRata-rata waktu penyembuhan Obat A: {mean_A:.2f} jam")
print(f"Rata-rata waktu penyembuhan Obat B: {mean_B:.2f} jam")
print(f"Rata-rata waktu penyembuhan Obat C: {mean_C:.2f} jam")
```

**Penjelasan Kode dan Hasil:**

1.  **`scipy.stats.f_oneway(obat_A, obat_B, obat_C)`:** Fungsi ini melakukan uji ANOVA satu arah. Ini mengambil kelompok-kelompok data yang berbeda sebagai argumen.
    * **Statistik F:** Ini adalah statistik uji untuk ANOVA. Ini adalah rasio varians antar kelompok (between-group variance) terhadap varians dalam kelompok (within-group variance). Statistik F yang lebih besar menunjukkan variabilitas yang lebih besar antara rata-rata kelompok relatif terhadap variabilitas di dalam kelompok.
    * **Nilai P (p-value):** Ini adalah probabilitas mengamati statistik F yang ekstrem atau lebih ekstrem dari yang dihitung, dengan asumsi hipotesis nol adalah benar.

2.  **Tingkat Signifikansi (`alpha`):** Ditetapkan pada 0.05 (5%) sesuai dengan soal.

3.  **Aturan Keputusan:**
    * Jika `nilai_p < alpha`, kita menolak hipotesis nol. Ini berarti ada cukup bukti untuk menyimpulkan bahwa setidaknya salah satu rata-rata berbeda.
    * Jika `nilai_p >= alpha`, kita gagal menolak hipotesis nol. Ini berarti tidak ada cukup bukti untuk menyimpulkan bahwa rata-rata tersebut berbeda.

**Kesimpulan:**

Berdasarkan uji ANOVA, nilai p (0.4578) lebih besar dari tingkat signifikansi (0.05). Oleh karena itu, kita **gagal menolak hipotesis nol**. Ini berarti bahwa tidak ada bukti statistik yang signifikan untuk menyimpulkan bahwa rata-rata waktu penyembuhan antara ketiga merek obat sakit kepala tersebut berbeda pada tingkat signifikansi 5%. Dengan kata lain, berdasarkan data ini, kita tidak dapat mengatakan bahwa satu obat sakit kepala menghasilkan rata-rata waktu penyembuhan yang secara signifikan berbeda dibandingkan dengan yang lain.

**STUDI KASUS 2**

image.png

Pak Rizal membudidayakan 3 jenis kentang berbeda di kebunnya. Untuk menilai apakah ada perbedaan rata-rata berat kentang, Pak Rizal mengambil 5 sampel dari masing-masing jenis kentang dan mencatat beratnya dalam gram:

Jenis 1: 254, 263, 241, 237, 251

Jenis 2: 234, 218, 235, 227, 216

Jenis 3: 200, 222, 197, 206, 204

Pak Rizal ingin mengetahui: Apakah perbedaan rata-rata berat antara ketiga jenis kentang ini signifikan secara statistik?
Diberikan bahwa populasi berdistribusi normal dan homogen, dan tingkat signifikansi adalah 1% (α = 0.01).

Hipotesis:
Hipotesis Nol (H₀): Tidak ada perbedaan rata-rata berat antara ketiga jenis kentang. (μ₁ = μ₂ = μ₃)

Hipotesis Alternatif (H₁): Setidaknya satu jenis kentang memiliki rata-rata berat yang berbeda.

```python
import scipy.stats as stats

# Data berat kentang per jenis
jenis1 = [254, 263, 241, 237, 251]
jenis2 = [234, 218, 235, 227, 216]
jenis3 = [200, 222, 197, 206, 204]

# Uji ANOVA satu arah
f_stat, p_val = stats.f_oneway(jenis1, jenis2, jenis3)

# Tingkat signifikansi
alpha = 0.01

print(f"Statistik F: {f_stat:.4f}")
print(f"Nilai P (p-value): {p_val:.6f}")
print(f"Tingkat Signifikansi (alpha): {alpha}")

# Keputusan
if p_val < alpha:
    print("\nKeputusan: Tolak Hipotesis Nol.")
    print("Kesimpulan: Terdapat perbedaan rata-rata berat yang signifikan antara ketiga jenis kentang.")
else:
    print("\nKeputusan: Gagal menolak Hipotesis Nol.")
    print("Kesimpulan: Tidak terdapat perbedaan rata-rata berat yang signifikan antara ketiga jenis kentang.")
```

**Penjelasan Hasil:**

Statistik F menunjukkan perbandingan variasi antar kelompok dengan variasi dalam kelompok.

Nilai p yang sangat kecil (< 0.01) menandakan bahwa kemungkinan mendapatkan hasil seperti ini secara kebetulan sangat kecil jika H₀ benar.

Karena p-value < α, maka H₀ ditolak.

**Kesimpulan Akhir:**

Berdasarkan uji ANOVA satu arah dan data berat kentang, terdapat perbedaan rata-rata yang signifikan secara statistik antara ketiga jenis kentang yang dibudidayakan oleh Pak Rizal pada taraf signifikansi 1%.

Dengan kata lain, setidaknya satu jenis kentang memiliki berat yang berbeda signifikan dari jenis lainnya.
